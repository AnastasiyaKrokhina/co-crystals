```python
import pandas as pd
```

```python
url11 = 'https://raw.githubusercontent.com/AnastasiyaKrokhina/co-crystals-by-a.krokhina/main/coformers1.csv'
url12 = 'https://raw.githubusercontent.com/AnastasiyaKrokhina/co-crystals-by-a.krokhina/main/coformers2.csv'
url21 = 'https://raw.githubusercontent.com/AnastasiyaKrokhina/co-crystals-by-a.krokhina/main/dataset1.csv'
url22 = 'https://raw.githubusercontent.com/AnastasiyaKrokhina/co-crystals-by-a.krokhina/main/dataset2.csv'
```

```python
df11 = pd.read_csv(url11)
df12 = pd.read_csv(url12)
df21 = pd.read_csv(url21)
df22 = pd.read_csv(url22)
```

```python
df11.head()
```

```python
df21.head()
```

```python
# how much unique values
len(df11.Identifier.append(df21.id).unique())
```

```python
df1 = pd.merge(df11[['Identifier', 'NAME']], df12[['Identifier', 'NAME']], how = 'left', on = 'Identifier').fillna('')
df1.head()
```

```python
df1.columns = ['id', 'smiles1', 'smiles2']
```

```python
df2 = pd.merge(df21[['id', 'smiles1']], df22[['id', 'smiles2']], how = 'left', on = 'id').fillna('')
df2.head()
```

```python
df = pd.concat([df1, df2], ignore_index=True).drop_duplicates(subset=['id'])
```

```python
from rdkit import Chem
from mordred import Calculator, descriptors
```

```python
calc = Calculator(descriptors, 
                  #ignore_3D=True
                  )
mol = Chem.MolFromSmiles('c1ccccc1')
calc(mol)[:3]
```

```python
# Calculate the descriptors for the first dataset (first co-former)
#dataset_mol1 = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/cocrystal_design/data/dataset_mol1.csv')
descriptors_mol1 =[]
for mol in df.smiles1.values:
    try:
        descriptors_mol1.append(calc(Chem.MolFromSmiles(mol)))
    except TypeError:
        descriptors_mol1.append('none')
# Save the dataset1 with the descriptors
descriptors1 = pd.DataFrame(descriptors_mol1)
df_mol1 = pd.concat((df, descriptors1), axis=1)
df_mol1
```
